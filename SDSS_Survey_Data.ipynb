```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
table_SDSS = pd.read_csv('M67.csv')
```

```python
center_x = 132.825
center_y = 11.8167

mask005in = ((table_SDSS['ra'] - center_x)**2 + (table_SDSS['dec'] - center_y)**2 < 0.05**2)

mask005out = ((table_SDSS['ra'] - center_x)**2 + (table_SDSS['dec'] - center_y)**2 > 0.05**2)

table_SDSS_005in = table_SDSS[mask005in]
table_SDSS_005out = table_SDSS[mask005out]
```

```python
u_g = table_SDSS['u'] - table_SDSS['g']
r_i = table_SDSS['r'] - table_SDSS['i']
r_z = table_SDSS['r'] - table_SDSS['z']
g_i = table_SDSS['g'] - table_SDSS['i']

fig, ax = plt.subplots(2,2)
fig.set_size_inches(11,8.5)

ax[0,0].scatter(table_SDSS['g'], u_g, color='red', s=1);
ax[0,0].set_ylabel("U-G")
ax[0,0].set_xlabel("G")

ax[0,1].scatter(r_i, table_SDSS['r'], color='green', s=1);
ax[0,1].set_xlabel("R-I")
ax[0,1].set_ylabel("R")

ax[1,0].scatter(u_g, r_i, color='blue', s=1);
ax[1,0].set_ylabel("R-I")
ax[1,0].set_xlabel("U-G")

ax[1,1].scatter(g_i, r_z, color='orange', s=1);
ax[1,1].set_ylabel("R-Z")
ax[1,1].set_xlabel("G-I")


ax[0,0].invert_yaxis();

ax[0,1].invert_yaxis();

ax[1,0].invert_yaxis();

ax[1,1].invert_yaxis();
ax[1,1].set_xlim(0,3.5);
```
